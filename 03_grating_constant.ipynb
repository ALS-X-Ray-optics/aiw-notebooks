Calculating the angles for a grating with a constant deflection angle.

```python
import numpy as np
import matplotlib.pyplot as plt

# grating groove density
g_lpm = 200e3; 

# complement of the included angle (theta' = 180-theta)
thetap_rad = 6*np.pi/180

Es_eV = np.linspace(200, 1000, 1000) # photon energy in eV
lambdas_m = 1239e-9/Es_eV # wavelength in m

# entrance angle (the parameter to be calculated)
alphap_rad = (thetap_rad**2-2*lambdas_m*g_lpm)/(2*thetap_rad)
betap_rad = thetap_rad-alphap_rad

plt.figure(figsize=(8, 6))
plt.plot(Es_eV, np.degrees(alphap_rad), label='Entrance angle (alpha\')')
plt.plot(Es_eV, np.degrees(betap_rad), label='Exit angle (beta\')')
plt.xlabel('Photon energy (eV)')
plt.ylabel('Angle (degrees)')
plt.title('Grating angles')
plt.legend()
plt.grid()
plt.xlim(200, 1000)
```
